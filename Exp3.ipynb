```python
# Student Performance Prediction
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

np.random.seed(42)
hours = np.random.randint(1, 11, 20)
marks = 5 * hours + 40 + np.random.normal(0, 2, 20)

print("Hours: ", hours)
print("Marks: ", marks,"\n")

hours_norm = (hours - np.mean(hours)) / np.std(hours)

w = 0
b = 0
learning_rate = 0.01
iterations = 1000

for i in range(1, iterations + 1):
    marks_pred = w * hours_norm + b
    error = marks_pred - marks

    dw = (2 / len(hours_norm)) * np.sum(error * hours_norm)
    db = (2 / len(hours_norm)) * np.sum(error)

    w -= learning_rate * dw
    b -= learning_rate * db

    mse = np.mean(error ** 2)

    if i % 100 == 0 or i == 1:
        print(f"Iteration {i}: Weight = {w:.5f}, Bias = {b:.5f}, MSE = {mse:.5f}")

predict_hours = 3
predict_hours_norm = (predict_hours - np.mean(hours)) / np.std(hours)
predicted_marks = w * predict_hours_norm + b

print("\nFinal Trained Weight:", round(w, 4))
print("Final Trained Bias:", round(b, 4))
print(f"\nPredicted Marks for {predict_hours} hours of study: {round(predicted_marks, 2)}")

print("\nActivation Function")
print("Sigmoid:", sigmoid(predicted_marks))
print("ReLU   :", relu(predicted_marks))
print("Tanh   :", tanh(predicted_marks))
```

```python
# Temperature Prediction
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

altitude = np.array([0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0])
temperature = np.array([28, 26, 23, 20, 17, 14, 11, 8, 5, 2])

print("Altitude (km):", altitude)
print("Temperature (°C):", temperature, "\n")

altitude_norm = (altitude - np.mean(altitude)) / np.std(altitude)

w = 0
b = 0
learning_rate = 0.01
iterations = 1000

for i in range(1, iterations + 1):
    temp_pred = w * altitude_norm + b
    error = temp_pred - temperature

    dw = (2 / len(altitude_norm)) * np.sum(error * altitude_norm)
    db = (2 / len(altitude_norm)) * np.sum(error)

    w -= learning_rate * dw
    b -= learning_rate * db

    mse = np.mean(error ** 2)

    if i % 100 == 0 or i == 1:
        print(f"Iteration {i}: Weight = {w:.5f}, Bias = {b:.5f}, MSE = {mse:.5f}")

new_altitude = 0.45
new_altitude_norm = (new_altitude - np.mean(altitude)) / np.std(altitude)
predicted_temp = w * new_altitude_norm + b

print("\nFinal Trained Weight:", round(w, 4))
print("Final Trained Bias:", round(b, 4))
print(f"Predicted temperature at {new_altitude} km: {round(predicted_temp, 2)} °C")

print("\nActivation Function")
print("Sigmoid:", sigmoid(predicted_temp))
print("ReLU   :", relu(predicted_temp))
print("Tanh   :", tanh(predicted_temp))
```

```python
# House Price Prediction
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

house_size = np.array([1, 2.5, 3.5, 4, 6, 7.5, 8.5, 9.5, 11, 14])
house_price = np.array([5, 6, 9, 10, 12, 15, 17, 20, 25, 30])

print("House Size (100s sq.ft):", house_size)
print("House Price (₹ Thousands):", house_price, "\n")

house_size_norm = (house_size - np.mean(house_size)) / np.std(house_size)

w = 0
b = 0
learning_rate = 0.002
iterations = 1000

for i in range(1, iterations + 1):
    price_pred = w * house_size_norm + b
    error = price_pred - house_price

    dw = (2 / len(house_size)) * np.sum(error * house_size_norm)
    db = (2 / len(house_size)) * np.sum(error)

    w -= learning_rate * dw
    b -= learning_rate * db

    mse = np.mean(np.square(error))

    if i % 100 == 0 or i == 1:
        print(f"Iteration {i}: Weight = {w:.5f}, Bias = {b:.5f}, MSE = {mse:.5f}")

new_size = 1
new_size_norm = (new_size - np.mean(house_size)) / np.std(house_size)
predicted_price = (w * new_size_norm) + b

print(f"\nFinal Weight: {round(w, 4)} and Bias: {round(b, 4)}")
print(f"Predicted House Price for {new_size} (100s sq.ft): ₹{round(predicted_price, 2)} Thousand")

print("\nActivation Function")
print("Sigmoid:", sigmoid(predicted_price))
print("ReLU   :", relu(predicted_price))
print("Tanh   :", tanh(predicted_price))
```
